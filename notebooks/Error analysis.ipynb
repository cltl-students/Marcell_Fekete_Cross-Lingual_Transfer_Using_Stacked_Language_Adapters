```python
import os
import random
import pandas as pd

random.seed(42)

pd.set_option("display.max_rows", None)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.width', None)
```

```python
generic_path = "..results/outputs/"
```

```python
def load_and_sample_df(language_pair, setup_name):
    filepath = os.path.join(generic_path, language_pair, setup_name, "output.tsv")
    df = pd.read_csv(filepath, sep="\t", index_col=0)
    df = df.filter(items=["Sentence", "Predicted", "Gold", "K_3_correct"])
    return df
```

```python
amharic_target_lang = load_and_sample_df("eng-amh", "target-lang").sample(100)
amharic_target_lang
```

```python
amharic_low_dttr = load_and_sample_df("eng-amh", "low-ttr").iloc[amharic_target_lang.index]
amharic_low_dttr
```
